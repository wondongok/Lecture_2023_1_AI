## AND / OR / XOR 이해하기 with keras

```python
from keras.models import Sequential
from keras.layers import InputLayer, Dense, Dropout, Activation
#from keras.optimizers import Adam, SGD
from tensorflow.keras.optimizers import Adam, SGD
import numpy as np
```

```python
x_data = [[0, 0],
          [0, 1],
          [1, 0],
          [1, 1]]
y_data = [[0],
          [0],
          [0],
          [1]]
x_data = np.array(x_data, dtype=np.float32)
y_data = np.array(y_data, dtype=np.float32)
```

```python
model = Sequential()
model.add(InputLayer(2))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
#optimizer = Adam(learning_rate=0.1) # learning rate
optimizer = SGD(learning_rate=0.1) # learning rate
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer=optimizer)
```

```python
model.summary()
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model.png', show_shapes=True)
```

```python
history = model.fit(x_data, y_data, batch_size=1, epochs=500)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], 'y', label='train loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
plt.plot(history.history['accuracy'], 'y', label='train acc')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

```python
#hypothesis = model.predict_proba(x_data)
hypothesis=model.predict(x_data)
print(hypothesis)
```
